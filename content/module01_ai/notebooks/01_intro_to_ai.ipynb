## Lesson 1: What is AI? — Agent Simulation

An intelligent agent perceives its environment and acts to achieve goals.

```python
class Agent:
    """A generic intelligent agent."""
    def __init__(self, name, sensors, actuators, goal):
        self.name = name
        self.sensors = sensors
        self.actuators = actuators
        self.goal = goal
        self.history = []
    
    def describe(self):
        return (
            f"Agent: {self.name}\n"
            f"  Sensors: {', '.join(self.sensors)}\n"
            f"  Actuators: {', '.join(self.actuators)}\n"
            f"  Goal: {self.goal}"
        )

# Create a drug discovery agent
drug_agent = Agent(
    "DrugDiscoveryAI",
    ["genomic_data", "protein_structures", "chemical_library", "clinical_trials"],
    ["candidate_molecules", "toxicity_predictions", "trial_recommendations"],
    "identify novel drug candidates for cancer treatment"
)

print(drug_agent.describe())
```

## Lesson 2: History of AI — Timeline Visualization

```python
milestones = [
    (1950, "Turing Test", "foundation"),
    (1956, "Dartmouth Conference", "foundation"),
    (1966, "ELIZA", "breakthrough"),
    (1974, "First AI Winter", "winter"),
    (1980, "Expert Systems Boom", "commercial"),
    (1987, "Second AI Winter", "winter"),
    (1997, "Deep Blue", "breakthrough"),
    (2006, "Deep Learning", "breakthrough"),
    (2012, "AlexNet", "breakthrough"),
    (2020, "AlphaFold", "breakthrough"),
    (2022, "ChatGPT", "commercial"),
]

color_map = {'foundation': 'blue', 'breakthrough': 'green', 'winter': 'red', 'commercial': 'orange'}

fig, ax = plt.subplots(figsize=(14, 3))
years = [m[0] for m in milestones]
colors = [color_map[m[2]] for m in milestones]

ax.scatter(years, [1]*len(years), c=colors, s=200, zorder=5)
ax.axhline(y=1, color='gray', linewidth=2)
ax.axvspan(1974, 1980, alpha=0.2, color='red', label='AI Winter')
ax.axvspan(1987, 1993, alpha=0.2, color='red')

for y, event, cat in milestones:
    ax.annotate(event, (y, 1.05), ha='center', fontsize=8, rotation=45)

ax.set_xlim(1945, 2027)
ax.set_ylim(0.5, 1.5)
ax.set_yticks([])
ax.set_xlabel('Year')
ax.set_title('AI History Timeline', fontsize=14, fontweight='bold')
ax.legend()
plt.tight_layout()
plt.show()
```

## Lesson 3: Types of AI — Classification Framework

```python
def classify_ai_system(description):
    """Classify an AI system by capability and functionality."""
    desc = description.lower()
    
    # All current systems are Narrow AI
    capability = "Narrow AI"
    
    # Check for learning/memory keywords
    memory_kw = ['learn', 'past', 'history', 'train', 'track', 'memory', 'experience']
    has_memory = any(kw in desc for kw in memory_kw)
    functionality = "Limited Memory" if has_memory else "Reactive Machine"
    
    return capability, functionality

systems = [
    "A chess engine that evaluates board positions",
    "A self-driving car that tracks nearby vehicles",
    "A language model trained on billions of documents",
    "A thermostat that reads temperature",
]

print("AI System Classification:")
for s in systems:
    cap, func = classify_ai_system(s)
    print(f"  {s[:55]:55s} → {cap:12s}, {func:20s}")
```

## Lesson 4: AI Paradigms — Rule-Based vs. ML

```python
# Rule-Based System
def wine_quality_rules(acidity, sugar, alcohol):
    if alcohol > 13 and acidity < 0.4:
        return 'High quality'
    elif alcohol > 12 and sugar < 2:
        return 'Medium quality'
    else:
        return 'Standard quality'

# ML System
wine = load_wine()
X_w, y_w = wine.data, wine.target
Xw_train, Xw_test, yw_train, yw_test = train_test_split(
    X_w, y_w, test_size=0.3, random_state=42
)

rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(Xw_train, yw_train)
rf_acc = accuracy_score(yw_test, rf.predict(Xw_test))

print("Paradigm Comparison:")
print(f"  Rule-Based: Returns quality based on 2 rules")
print(f"  ML (Random Forest): {rf_acc:.1%} accuracy on wine classification")
```

## Lesson 5: AI Applications — Recommendation System

```python
ratings = np.array([
    [5, 4, 0, 0, 1],
    [4, 5, 0, 0, 2],
    [0, 0, 5, 4, 0],
    [0, 0, 4, 5, 3],
    [1, 2, 0, 3, 5],
])

def recommend(ratings, user_id):
    target = ratings[user_id]
    similarities = []
    for i in range(len(ratings)):
        if i == user_id:
            continue
        common = (target > 0) & (ratings[i] > 0)
        if common.sum() == 0:
            sim = 0
        else:
            diff = np.abs(target[common] - ratings[i][common])
            sim = 1 / (1 + diff.mean())
        similarities.append((i, sim))
    
    most_similar = max(similarities, key=lambda x: x[1])[0]
    recs = [j for j in range(ratings.shape[1])
            if target[j] == 0 and ratings[most_similar][j] >= 4]
    return recs

print("Recommendation System (Collaborative Filtering):")
for uid in range(5):
    recs = recommend(ratings, uid)
    if recs:
        print(f"  User {uid}: Recommend items {recs}")
    else:
        print(f"  User {uid}: No new recommendations")
```

## Lesson 6: AI in Biotech — Virtual Screening

```python
from sklearn.datasets import make_classification

X, y = make_classification(n_samples=2000, n_features=20, n_informative=10,
                           weights=[0.95, 0.05], random_state=42)
Xs_train, Xs_test, ys_train, ys_test = train_test_split(X, y, test_size=0.8, random_state=42)

model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(Xs_train, ys_train)
probs = model.predict_proba(Xs_test)[:, 1]

top_10_indices = np.argsort(probs)[-10:][::-1]
print("Virtual Screening - Top 10 Candidates:")
for i, idx in enumerate(top_10_indices, 1):
    print(f"  {i}. Compound {idx}: probability = {probs[idx]:.1%}, "
          f"actual = {'Active' if ys_test[idx] else 'Inactive'}")
print(f"\nEnrichment: {ys_test[top_10_indices].sum()}/10 active in top 10 "
      f"(vs. {ys_test.mean():.1%} baseline)")
```

## Lesson 7: AI in SaaS — Churn Prediction

```python
np.random.seed(42)
n = 2000
data = pd.DataFrame({
    'tenure': np.random.randint(1, 60, n),
    'logins': np.random.randint(0, 30, n),
    'tickets': np.random.randint(0, 15, n),
    'features': np.random.randint(0, 20, n),
    'spend': np.random.uniform(10, 200, n),
})
data['churn'] = ((data['logins'] < 3) | (data['tickets'] > 10)).astype(int)

X = data[['tenure', 'logins', 'tickets', 'features', 'spend']]
y = data['churn']
X_tr, X_te, y_tr, y_te = train_test_split(X, y, test_size=0.3, random_state=42)

rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_tr, y_tr)
y_pred = rf.predict(X_te)

print("Churn Prediction Results:")
print(classification_report(y_te, y_pred))
print(f"\nTop 3 churn drivers:")
imp = pd.DataFrame({'feature': X.columns, 'importance': rf.feature_importances_})
print(imp.sort_values('importance', ascending=False).head(3).to_string(index=False))
```
