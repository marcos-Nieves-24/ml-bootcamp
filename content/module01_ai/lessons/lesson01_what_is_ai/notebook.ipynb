## Intelligent Agent Simulation

Let us simulate a simple intelligent agent: a **thermostat**. This is the simplest form of an AI agent — it perceives temperature, compares it to a set point, and acts to maintain comfort.

```python
def thermostat_agent(current_temp, set_point, threshold=1.0):
    """
    A simple thermostat agent.
    
    Parameters:
    - current_temp: current room temperature (sensor input)
    - set_point: desired temperature (goal)
    - threshold: acceptable deviation
    
    Returns:
    - action: 'heat', 'cool', or 'idle'
    """
    diff = current_temp - set_point
    
    if diff < -threshold:
        return 'heat'
    elif diff > threshold:
        return 'cool'
    else:
        return 'idle'


# Test the agent
temperatures = [15.0, 18.5, 21.0, 23.5, 28.0]
set_point = 21.0

print(f"Set point: {set_point}°C")
print(f"Threshold: ±1.0°C")
print("-" * 40)
for temp in temperatures:
    action = thermostat_agent(temp, set_point)
    print(f"Temperature: {temp}°C → Action: {action}")
```

### Interpreting the Results

The thermostat agent uses a simple rule-based system:
- **Sensor**: temperature reading
- **Environment**: the room
- **Actuator**: heating/cooling system
- **Performance measure**: keeping temperature close to the set point

This is a **rational agent** — it acts to achieve a goal.

## Implementing a Generic Agent Class

Let us generalize the agent concept using object-oriented programming.

```python
class Agent:
    """A generic intelligent agent."""
    
    def __init__(self, name, sensors, actuators, goal):
        self.name = name
        self.sensors = sensors
        self.actuators = actuators
        self.goal = goal
        self.history = []
    
    def perceive(self, observation):
        """Receive a sensor observation."""
        self.history.append(observation)
    
    def act(self, action):
        """Execute an action."""
        print(f"{self.name} performs: {action}")
    
    def describe(self):
        """Return a description of the agent."""
        sensors_str = ", ".join(self.sensors)
        actuators_str = ", ".join(self.actuators)
        return (
            f"Agent '{self.name}'\n"
            f"  Sensors: {sensors_str}\n"
            f"  Actuators: {actuators_str}\n"
            f"  Goal: {self.goal}\n"
            f"  Percepts so far: {len(self.history)}"
        )


# Create a Spotify-like recommendation agent
recommender = Agent(
    name="MusicRecommender",
    sensors=["listening_history", "likes", "skips", "playlists"],
    actuators=["song_suggestions", "playlist_generation"],
    goal="maximize user engagement"
)

print(recommender.describe())
```

```python
# Exercise 1: Variable threshold thermostat
def smart_thermostat(current_temp, set_point, hour):
    """Thermostat with time-dependent threshold."""
    if 6 <= hour <= 22:
        threshold = 0.5  # Tighter control during waking hours
    else:
        threshold = 2.0  # Wider tolerance at night
    
    diff = current_temp - set_point
    if diff < -threshold:
        return 'heat'
    elif diff > threshold:
        return 'cool'
    else:
        return 'idle'


# Test
for hour in [3, 10, 14, 23]:
    action = smart_thermostat(18.0, 21.0, hour)
    print(f"Hour {hour}: {action}")
```
