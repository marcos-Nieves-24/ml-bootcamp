# Lesson 10: Tuples

Learn about immutable sequences and tuple unpacking.

## Creating Tuples

```python
import math  # noqa: F401

empty = ()
single = (42,)
pair = (10, 20)
triple = 1, 2, 3

print(f"empty: {empty}, type: {type(empty)}")
print(f"single: {single}, type: {type(single)}")
print(f"pair: {pair}")
print(f"triple: {triple}")
```

## Immutability

```python
t = (1, 2, 3)
print(f"Original: {t}")

# Uncomment to see error:
# t[0] = 99  # TypeError

# But we can create new tuples
t2 = (99,) + t[1:]
print(f"New tuple: {t2}")
```

## Tuple Unpacking

```python
point = (3, 7)
x, y = point
print(f"x={x}, y={y}")

# Swapping
a, b = 10, 20
a, b = b, a
print(f"After swap: a={a}, b={b}")

# Extended unpacking
first, *middle, last = (1, 2, 3, 4, 5)
print(f"first={first}, middle={middle}, last={last}")
```

## Multiple Return Values

```python
def min_max(numbers):
    return min(numbers), max(numbers)

data = [3, 1, 7, 2, 9, 4]
mn, mx = min_max(data)
print(f"Data: {data}")
print(f"Min: {mn}, Max: {mx}")
```

## Tuples as Dictionary Keys

```python
locations = {
    (40.7128, -74.0060): "New York",
    (34.0522, -118.2437): "Los Angeles",
    (41.8781, -87.6298): "Chicago",
}

for coords, city in locations.items():
    print(f"{city}: {coords}")
```

## Biotechnology Example

```python
variants = [
    ("chr1", 123456, "A", "G", 99),
    ("chr1", 789012, "C", "T", 85),
    ("chrX", 345678, "G", "A", 92),
]

for chrom, pos, ref, alt, qual in variants:
    if qual > 90:
        print(f"High-quality: {chrom}:{pos} {ref}→{alt} (Q{qual})")
```

## SaaS Example

```python
events = [
    (1700000000, "user_001", "login", ("127.0.0.1", "Chrome")),
    (1700000050, "user_042", "purchase", ("premium", 49.99)),
]

for ts, uid, event_type, meta in events:
    if event_type == "purchase":
        product, amount = meta
        print(f"{uid} bought {product} for ${amount:.2f}")
```

## Summary
- Tuples: ordered, immutable sequences
- `(42,)` for single element (comma required!)
- Unpacking: `x, y = (3, 4)`
- Hashable → dictionary keys
- Use for fixed data and multiple returns
