# Lesson 14: Pandas

Learn data manipulation with Pandas DataFrames and Series.

```python
import pandas as pd
import numpy as np
```

## Creating DataFrames

```python
df = pd.DataFrame({
    "Gene": ["BRCA1", "TP53", "EGFR", "MYC", "KRAS"],
    "Expression": [2.5, -1.2, 3.8, 0.9, -2.1],
    "P_Value": [0.001, 0.08, 0.003, 0.45, 0.02],
    "Chromosome": [17, 17, 7, 8, 12]
})
print(df)
```

## Exploring Data

```python
print("Head:")
print(df.head())
print("\nInfo:")
print(df.info())
print("\nDescribe:")
print(df.describe())
```

## Selecting and Filtering

```python
print(f"Column: {df['Gene']}")
print(f"\nSignificant (p < 0.05):")
print(df[df["P_Value"] < 0.05])
print(f"\nRow 0: {df.loc[0]}")
```

## Adding Columns

```python
df["Significant"] = df["P_Value"] < 0.05
df["Abs_Expression"] = df["Expression"].abs()
print(df)
```

## GroupBy

```python
sales = pd.DataFrame({
    "Product": ["A", "B", "A", "C", "B", "A"],
    "Region": ["North", "South", "South", "North", "North", "South"],
    "Revenue": [100, 200, 150, 300, 250, 120]
})

print("By product:")
print(sales.groupby("Product")["Revenue"].sum())
print("\nBy product and region:")
print(sales.groupby(["Product", "Region"])["Revenue"].mean())
```

## Merging

```python
customers = pd.DataFrame({"ID": [1, 2, 3], "Name": ["Alice", "Bob", "Charlie"]})
orders = pd.DataFrame({"OrderID": [101, 102, 103], "CustomerID": [1, 1, 3], "Amount": [50, 75, 100]})

merged = pd.merge(customers, orders, left_on="ID", right_on="CustomerID")
print(merged)
```

## Apply Functions

```python
df["Expression_Rounded"] = df["Expression"].apply(lambda x: round(x, 1))
df["Direction"] = df["Expression"].apply(lambda x: "Up" if x > 0 else "Down")
print(df[["Gene", "Expression", "Direction"]])
```

## Biotechnology Example

```python
np.random.seed(42)
clinical = pd.DataFrame({
    "PatientID": [f"P-{i:04d}" for i in range(1, 21)],
    "Age": np.random.randint(18, 85, 20),
    "Treatment": np.random.choice(["Drug", "Placebo"], 20),
    "Response": np.random.choice(["Responder", "Non-responder"], 20, p=[0.6, 0.4]),
})

print(clinical.groupby("Treatment")["Response"].value_counts())
```

## Summary
- DataFrame: 2D labeled table
- Series: 1D labeled column
- groupby: split-apply-combine
- merge: SQL-like joins
- apply: transform values
- Foundation for data analysis
