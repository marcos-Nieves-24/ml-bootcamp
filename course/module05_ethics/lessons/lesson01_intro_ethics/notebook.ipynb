## 1. The Five Principles Framework

Floridi and Cowls (2019) identified five core principles that synthesize over 50 AI ethics guidelines:

| Principle | Definition | Key Question |
|-----------|------------|-------------|
| Beneficence | Do good | Does the system promote well-being? |
| Non-maleficence | Do no harm | Could the system cause harm? |
| Autonomy | Respect human agency | Are humans in control? |
| Justice | Be fair | Are benefits and burdens distributed equitably? |
| Explicability | Be transparent | Can decisions be understood and challenged? |

These principles often conflict. There is no universal resolution.

## 2. Thought Experiment: Predictive Policing

Consider a predictive policing system that uses historical crime data to allocate patrols.

Analyze using the five principles:

- **Beneficence**: Could reduce crime, improve public safety
- **Non-maleficence**: Historical data reflects biased policing; may perpetuate over-policing
- **Autonomy**: Officers may defer to algorithm rather than using judgment
- **Justice**: If data is biased, system unfairly targets already over-policed communities
- **Explicability**: Citizens may not know they are policed by algorithm
