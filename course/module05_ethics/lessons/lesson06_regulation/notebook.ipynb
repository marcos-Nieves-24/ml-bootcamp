## 1. EU AI Act Risk Classification Tool

We create a decision tool that classifies AI systems under the EU AI Act.

```python
def classify_eu_ai_act(system_description):
    """
    Simple rule-based classification tool for EU AI Act risk levels.
    Returns risk level and key requirements.
    """
    high_risk_categories = [
        'medical device', 'hiring', 'credit scoring', 'law enforcement',
        'critical infrastructure', 'education', 'migration', 'biometric identification'
    ]
    prohibited_practices = [
        'social scoring', 'real-time biometric surveillance',
        'manipulative', 'exploiting vulnerabilities'
    ]
    
    desc_lower = system_description.lower()
    
    for practice in prohibited_practices:
        if practice in desc_lower:
            return {
                'Risk Level': 'Unacceptable (Prohibited)',
                'Requirements': 'None - system cannot be deployed'
            }
    
    for category in high_risk_categories:
        if category in desc_lower:
            return {
                'Risk Level': 'High-Risk',
                'Requirements': ("Conformity assessment, risk management, "
                               "human oversight, transparency, documentation, "
                               "data governance, accuracy/robustness/cybersecurity")
            }
    
    if any(term in desc_lower for term in ['chatbot', 'deepfake', 'emotion recognition']):
        return {
            'Risk Level': 'Limited Risk',
            'Requirements': 'Transparency obligations (disclose AI interaction)'
        }
    
    return {
        'Risk Level': 'Minimal Risk',
        'Requirements': 'No specific obligations under AI Act'
    }
```

```python
systems = [
    "AI system that screens job applications and ranks candidates for hiring",
    "AI-powered chatbot for customer service in an e-commerce platform",
    "Real-time biometric surveillance system in public spaces",
    "AI system that analyzes medical images to detect tumors",
    "AI system that recommends products to users based on browsing history",
    "AI system that assigns a social credit score to citizens"
]

classifications = []
for s in systems:
    result = classify_eu_ai_act(s)
    result['System'] = s
    classifications.append(result)

df = pd.DataFrame(classifications)[['System', 'Risk Level', 'Requirements']]
print("EU AI Act Classification Results:")
for i, row in df.iterrows():
    print(f"\n{row['System']}")
    print(f"  Risk Level: {row['Risk Level']}")
    print(f"  Requirements: {row['Requirements']}")
```

## 2. Algorithmic Audit Checklist

We create an audit checklist framework and apply it to a hypothetical hiring AI.

```python
audit_checklist = pd.DataFrame({
    'Audit Area': [
        'Data', 'Data', 'Data',
        'Model', 'Model', 'Model',
        'Transparency', 'Transparency',
        'Governance', 'Governance'
    ],
    'Question': [
        'Is the training data representative of the target population?',
        'Are there known biases in the data collection process?',
        'Is the label quality consistent across demographic groups?',
        'Does the model achieve comparable accuracy across all demographic groups?',
        'Has the model been tested for robustness (adversarial examples, distribution shift)?',
        'Are fairness metrics (DP, EO, EOdds) computed and monitored?',
        'Is there documentation explaining how the model works to affected users?',
        'Is there a mechanism for users to appeal or contest decisions?',
        'Is there an internal ethics review process for new AI products?',
        'Is there a clear allocation of responsibility for AI outcomes?'
    ],
    'Status': ['Not Checked'] * 10
})

print("Algorithmic Audit Checklist:")
print(audit_checklist.to_string(index=False))
```

## 3. Governance Simulation

We simulate a simple ethics board review process.

```python
def ethics_board_review(system_name, risk_score, fairness_score, transparency_score):
    """
    Simulate an AI ethics board review.
    risk_score: 0-100 (higher = more risky)
    fairness_score: 0-100 (higher = more fair)
    transparency_score: 0-100 (higher = more transparent)
    """
    print(f"=== Ethics Board Review: {system_name} ===")
    print(f"Risk Score: {risk_score}/100")
    print(f"Fairness Score: {fairness_score}/100")
    print(f"Transparency Score: {transparency_score}/100")
    
    issues = []
    if risk_score > 70:
        issues.append("HIGH RISK: Requires detailed risk mitigation plan")
    if fairness_score < 60:
        issues.append("FAIRNESS CONCERN: Bias testing and mitigation required before deployment")
    if transparency_score < 50:
        issues.append("TRANSPARENCY GAP: User-facing documentation needed")
    
    if not issues:
        print("Verdict: APPROVED - System meets ethical standards")
    elif len(issues) <= 2:
        print("Verdict: CONDITIONALLY APPROVED - Issues must be addressed:")
        for issue in issues:
            print(f"  - {issue}")
    else:
        print("Verdict: BLOCKED - Deployment not authorized until issues resolved:")
        for issue in issues:
            print(f"  - {issue}")
    return len(issues)

# Simulate reviews
ethics_board_review("Diagnostic AI for Cancer Detection", 65, 72, 80)
print()
ethics_board_review("Automated Resume Screening", 85, 45, 30)
print()
ethics_board_review("Product Recommendation System", 25, 90, 55)
```
