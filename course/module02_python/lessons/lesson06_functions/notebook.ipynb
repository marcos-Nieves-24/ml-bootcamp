# Lesson 6: Functions

Learn to write reusable code blocks with functions.

## Defining and Calling Functions

```python
def greet(name):
    """Return a greeting message."""
    return f"Hello, {name}!"

message = greet("Alice")
print(message)
```

## Parameters and Arguments

```python
def describe_patient(name, age, diagnosis="Unknown"):
    """Generate a patient summary."""
    return f"{name}, {age} yrs, Diagnosis: {diagnosis}"

# Positional arguments
print(describe_patient("Alice", 45, "Hypertension"))

# Keyword arguments
print(describe_patient(age=30, name="Bob"))

# Default parameter
print(describe_patient("Charlie", 25))
```

## Return Values

```python
def analyze(vital_signs):
    """Analyze vital signs and return multiple values."""
    hr = vital_signs["heart_rate"]
    bp_sys = vital_signs["systolic_bp"]
    bp_dia = vital_signs["diastolic_bp"]
    
    is_tachycardic = hr > 100
    is_hypertensive = bp_sys > 130 or bp_dia > 85
    
    return is_tachycardic, is_hypertensive

patient = {"heart_rate": 110, "systolic_bp": 140, "diastolic_bp": 90}
tachy, hyper = analyze(patient)
print(f"Tachycardic: {tachy}, Hypertensive: {hyper}")
```

## Lambda Functions

```python
# Lambda for squaring
square = lambda x: x ** 2
print(f"Square of 5: {square(5)}")

# Lambda with map
numbers = [1, 2, 3, 4, 5]
cubes = list(map(lambda x: x ** 3, numbers))
print(f"Cubes: {cubes}")

# Lambda with filter
evens = list(filter(lambda x: x % 2 == 0, numbers))
print(f"Evens: {evens}")
```

## Scope

```python
x = "global"

def test_scope():
    y = "local"
    print(f"Inside: x={x}, y={y}")

test_scope()
print(f"Outside: x={x}")
```

## Biotechnology Example: DNA Analysis

```python
def gc_content(sequence):
    """Calculate GC content percentage."""
    seq = sequence.upper()
    gc = seq.count("G") + seq.count("C")
    return (gc / len(seq)) * 100

def reverse_complement(sequence):
    """Return reverse complement of DNA."""
    comp = {"A": "T", "T": "A", "C": "G", "G": "C"}
    return "".join(comp[b] for b in reversed(sequence.upper()))

dna = "ATCGATCGATCG"
print(f"Sequence: {dna}")
print(f"GC content: {gc_content(dna):.1f}%")
print(f"Reverse complement: {reverse_complement(dna)}")
```

## SaaS Example: Customer Metrics

```python
def calculate_arpu(revenue, users):
    return revenue / users if users > 0 else 0.0

def churn_rate(lost, total):
    return (lost / total) * 100 if total > 0 else 0.0

revenue, users, lost = 125000, 15420, 540
print(f"ARPU: ${calculate_arpu(revenue, users):.2f}")
print(f"Churn Rate: {churn_rate(lost, users):.1f}%")
```

## Summary
- Functions are defined with `def` and called with `()`
- Parameters receive arguments; `return` sends output
- Lambda functions are anonymous single-expression functions
- Local scope vs global scope
- Docstrings document function purpose
