# Lesson 13: NumPy

Learn the foundation of scientific computing in Python — NumPy arrays.

```python
import numpy as np
```

## Creating Arrays

```python
arr1d = np.array([1, 2, 3, 4, 5])
arr2d = np.array([[1, 2, 3], [4, 5, 6]])
zeros = np.zeros((3, 4))
ones = np.ones((2, 3))
eye = np.eye(3)
arange = np.arange(0, 10, 2)
linspace = np.linspace(0, 1, 5)
random = np.random.randn(3, 3)

print(f"1D: {arr1d}")
print(f"2D:\n{arr2d}")
print(f"Zeros shape: {zeros.shape}")
```

## Array Attributes

```python
arr = np.array([[1, 2, 3], [4, 5, 6]])
print(f"Array:\n{arr}")
print(f"Shape: {arr.shape}")
print(f"Dimensions: {arr.ndim}")
print(f"Size: {arr.size}")
print(f"Dtype: {arr.dtype}")
```

## Indexing and Slicing

```python
matrix = np.arange(9).reshape(3, 3)
print(f"Matrix:\n{matrix}")
print(f"Element [1,2]: {matrix[1, 2]}")
print(f"First row: {matrix[0]}")
print(f"Last col: {matrix[:, -1]}")
print(f"Submatrix:\n{matrix[:2, :2]}")
print(f"Boolean: {matrix[matrix > 4]}")
```

## Universal Functions

```python
arr = np.array([1, 4, 9, 16, 25])
print(f"Original: {arr}")
print(f"sqrt: {np.sqrt(arr)}")
print(f"log: {np.log(arr)}")
print(f"+ 10: {arr + 10}")
print(f"* 2: {arr * 2}")
```

## Broadcasting

```python
matrix = np.ones((4, 3))
vector = np.array([1, 2, 3])
result = matrix + vector
print(f"Matrix:\n{matrix}")
print(f"Vector: {vector}")
print(f"Result:\n{result}")
```

## Linear Algebra

```python
a = np.array([[1, 2], [3, 4]])
b = np.array([[5, 6], [7, 8]])
print(f"A @ B:\n{a @ b}")
print(f"Inverse A:\n{np.linalg.inv(a)}")
print(f"Det A: {np.linalg.det(a)}")
```

## Biotechnology Example

```python
np.random.seed(42)
expression = np.random.randn(5, 4) * 2 + 5
genes = ["BRCA1", "TP53", "EGFR", "MYC", "KRAS"]

means = expression.mean(axis=1).reshape(-1, 1)
stds = expression.std(axis=1).reshape(-1, 1)
z_scores = (expression - means) / stds

for i, gene in enumerate(genes):
    print(f"{gene}: mean={expression[i].mean():.2f}, z-scores={z_scores[i].round(2)}")
```

## SaaS Example

```python
np.random.seed(42)
dau = np.random.randint(1000, 5000, size=(7, 3))
products = ["Dashboard", "Analytics", "Reports"]

for i, p in enumerate(products):
    print(f"{p}: {dau[:, i].mean():.0f} ± {dau[:, i].std():.0f}")

corr = np.corrcoef(dau.T)
print(f"Correlation:\n{corr}")
```

## Summary
- ndarray: fast, homogeneous, multidimensional
- Vectorized operations = C speed
- Broadcasting: flexible shape operations
- Universal functions: element-wise math
- Linear algebra: @, linalg
- Foundation for Pandas, scikit-learn
