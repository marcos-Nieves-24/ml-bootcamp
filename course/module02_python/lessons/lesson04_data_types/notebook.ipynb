# Lesson 4: Data Types

Explore Python's primitive data types: int, float, str, and bool.

## The Four Primitive Types

```python
import math  # noqa: F401

gene_count = 25000          # int
mutation_rate = 0.0015      # float
gene_name = "TP53"         # str
is_cancer_associated = True # bool

print(f"{gene_count} is {type(gene_count)}")
print(f"{mutation_rate} is {type(mutation_rate)}")
print(f"{gene_name} is {type(gene_name)}")
print(f"{is_cancer_associated} is {type(is_cancer_associated)}")
```

## Type Conversion

```python
# Explicit type conversion
count_str = "1500"
count_int = int(count_str)
count_float = float(count_int)

print(f"Original: {count_str} ({type(count_str)})")
print(f"As int: {count_int} ({type(count_int)})")
print(f"As float: {count_float} ({type(count_float)})")
```

```python
# int() truncates, doesn't round
print(f"int(3.999) = {int(3.999)}")
print(f"int(-3.5) = {int(-3.5)}")
```

## Boolean Conversion (Truthy/Falsy)

```python
values = [0, 1, -1, 0.0, 3.14, "", "hello", [], [1,2], None]
for v in values:
    print(f"bool({v!r:>8}) = {bool(v)}")
```

## String Operations

```python
dna = "ATCGATCG"
print(f"Length: {len(dna)}")
print(f"First 3 bases: {dna[:3]}")
print(f"GC count: {dna.count('G') + dna.count('C')}")
print(f"Complement: {dna.replace('A','t').replace('T','a').replace('C','g').replace('G','c').upper()}")
```

## Float Precision

```python
print(f"0.1 + 0.2 = {0.1 + 0.2}")
print(f"0.1 + 0.2 == 0.3: {0.1 + 0.2 == 0.3}")
print(f"Safe comparison: {abs((0.1 + 0.2) - 0.3) < 1e-10}")
```

## Type Errors

```python
# This will raise TypeError
try:
    result = "The answer is " + 42
except TypeError as e:
    print(f"Error: {e}")
    # Fix:
    result = "The answer is " + str(42)
    print(f"Fixed: {result}")
```

## Biotechnology Example

```python
gene_name = "EGFR"
expression_ratio = 2.45
p_value = 0.003
is_significant = p_value < 0.05
sample_count = 48

print(f"Gene: {gene_name}")
print(f"Expression ratio: {expression_ratio:.2f}")
print(f"p-value: {p_value} (significant: {is_significant})")
print(f"Sample size: {sample_count}")
```

## SaaS Example

```python
customer = "Acme Corp"
users = 342
revenue = 45900.75
is_premium = True
discount = 0.15

annual = revenue * 12
discounted = annual * (1 - discount)
print(f"Customer: {customer} (premium: {is_premium})")
print(f"Users: {users}")
print(f"Annual revenue: ${annual:,.2f}")
print(f"After discount: ${discounted:,.2f}")
```

## Summary
- int: whole numbers, float: decimal numbers, str: text, bool: True/False
- Use `type()` to check, `int()`/`float()`/`str()`/`bool()` to convert
- `input()` returns strings — convert as needed
- Be aware of float precision limitations
- All primitive types are immutable
