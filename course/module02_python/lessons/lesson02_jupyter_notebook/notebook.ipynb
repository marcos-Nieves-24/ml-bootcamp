## Python Code Cells

Code cells execute Python code. The output appears below the cell.

```python
import math  # noqa: F401

print("Hello from Jupyter Notebook!")
print("This is a code cell.")
```

```python
# Variables defined in one cell are available in subsequent cells
message = "Variables persist across cells"
number = 42
print(message)
print(f"The answer is {number}")
```

## Interactive Computing

You can use Jupyter for calculations and data exploration.

```python
# Using Python as a calculator
print(f"Sum: {10 + 20}")
print(f"Product: {10 * 20}")
print(f"Power: {2 ** 10}")
print(f"Division: {100 / 3:.2f}")
```

```python
# Creating and displaying a list
fruits = ["apple", "banana", "cherry", "date"]
for i, fruit in enumerate(fruits, 1):
    print(f"{i}. {fruit}")
```
