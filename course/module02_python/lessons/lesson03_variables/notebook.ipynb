# Lesson 3: Variables

Learn how to store and manage data using variables in Python.

## Basic Variable Assignment

```python
# Assign values to variables
gene = "BRCA1"
expression_level = 2.5
is_mutated = True

print("Gene:", gene)
print("Expression:", expression_level)
print("Mutated:", is_mutated)
```

## Dynamic Typing

```python
# Python infers the type from the value
x = 42
print(f"x = {x}, type = {type(x)}")

x = 3.14159
print(f"x = {x}, type = {type(x)}")

x = "Hello"
print(f"x = {x}, type = {type(x)}")

x = [1, 2, 3]
print(f"x = {x}, type = {type(x)}")
```

## Multiple Assignment

```python
# Tuple unpacking
a, b, c = 10, 20, 30
print(f"a={a}, b={b}, c={c}")

# Same value to multiple variables
x = y = z = 0
print(f"x={x}, y={y}, z={z}")

# Swapping
a, b = b, a
print(f"After swap: a={a}, b={b}")
```

## User Input

```python
# input() always returns a string
name = input("Enter your name: ")
age = int(input("Enter your age: "))
height = float(input("Enter your height in meters: "))

print(f"Hello {name}, you are {age} years old and {height}m tall.")
```

## Biotechnology Example: Patient Records

```python
patient_id = "P-0042"
age = 45
heart_rate = 72
systolic_bp = 120
diastolic_bp = 80
diagnosis = "Hypertension"

print(f"Patient: {patient_id}")
print(f"Age: {age} years")
print(f"Heart Rate: {heart_rate} bpm")
print(f"Blood Pressure: {systolic_bp}/{diastolic_bp} mmHg")
print(f"Diagnosis: {diagnosis}")
```

## SaaS Example: Business Metrics

```python
company = "DataCloud Inc."
revenue = 125000.50
users = 15420
churn_rate = 0.035
arpu = revenue / users

print(f"Company: {company}")
print(f"Monthly Revenue: ${revenue:,.2f}")
print(f"Active Users: {users:,}")
print(f"Churn Rate: {churn_rate:.1%}")
print(f"ARPU: ${arpu:.2f}")
```

## Summary

- Variables store data in named memory locations
- Python is dynamically typed — type is inferred from value
- Use `snake_case` names (PEP 8)
- `input()` always returns a string
- Multiple assignment and tuple unpacking are powerful Python features
