# Lesson 8: Conditionals

Learn to make decisions in Python using if/elif/else, truthy/falsy, and match statements.

## Basic if/elif/else

```python
import math  # noqa: F401

bmi = 26.3
if bmi < 18.5:
    category = "Underweight"
elif bmi < 25:
    category = "Normal"
elif bmi < 30:
    category = "Overweight"
else:
    category = "Obese"
print(f"BMI: {bmi:.1f} - {category}")
```

## Truthy and Falsy

```python
test_vals = [0, 1, "", "hello", [], [1,2], None, False, True]
for val in test_vals:
    if val:
        print(f"  Truthy: {val!r}")
    else:
        print(f"Falsy:   {val!r}")
```

## Ternary (Conditional Expression)

```python
age = 20
status = "Adult" if age >= 18 else "Minor"
print(f"Age {age} -> {status}")

# Equivalent using if/else
if age >= 18:
    status = "Adult"
else:
    status = "Minor"
print(f"Age {age} -> {status}")
```

## Match Statement (Python 3.10+)

```python
def describe_mutation(mut_type):
    match mut_type:
        case "missense":
            return "Single amino acid change"
        case "nonsense":
            return "Premature stop codon"
        case "frameshift":
            return "Reading frame shift"
        case "silent":
            return "No amino acid change"
        case _:
            return "Unknown type"

for m in ["missense", "frameshift", "unknown"]:
    print(f"{m}: {describe_mutation(m)}")
```

## Nested Conditionals

```python
x = 10
y = 5
if x > 0:
    if y > 0:
        print("Both positive")
    else:
        print("x positive, y non-positive")
else:
    print("x is non-positive")
```

## Biotechnology Example

```python
def assess_patient(sbp, dbp, hr, glucose):
    if sbp < 120 and dbp < 80:
        bp = "Normal"
    elif sbp < 130 and dbp < 80:
        bp = "Elevated"
    elif sbp < 140 or dbp < 90:
        bp = "Stage 1 HTN"
    else:
        bp = "Stage 2 HTN"
    
    hr_status = "Normal" if 60 <= hr <= 100 else "Abnormal"
    glucose_status = "Normal" if 70 <= glucose <= 126 else "Abnormal"
    
    emergency = bp == "Stage 2 HTN" or hr_status == "Abnormal" or glucose_status == "Abnormal"
    return bp, hr_status, glucose_status, emergency

result = assess_patient(145, 95, 110, 140)
print(f"BP: {result[0]}, HR: {result[1]}, Glucose: {result[2]}, Emergency: {result[3]}")
```

## SaaS Example

```python
def calculate_discount(tier, spend, years):
    match tier:
        case "enterprise":
            base = 0.20
        case "professional":
            base = 0.10
        case _:
            base = 0.05
    
    loyalty = 0.05 if years >= 3 else (0.02 if years >= 1 else 0)
    volume = 0.10 if spend > 50000 else (0.05 if spend > 10000 else 0)
    
    return min(base + loyalty + volume, 0.30)

discount = calculate_discount("professional", 25000, 4)
print(f"Discount: {discount:.0%}")
```

## Summary
- `if`/`elif`/`else` for decision making
- Truthy: non-zero, non-empty values
- Falsy: None, False, 0, empty sequences
- Ternary for simple inline conditions
- Match (3.10+) for pattern matching
- Use guard clauses to avoid deep nesting
