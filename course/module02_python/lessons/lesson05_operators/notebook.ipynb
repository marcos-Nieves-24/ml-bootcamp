# Lesson 5: Operators

Learn about arithmetic, comparison, logical, and assignment operators in Python.

## Arithmetic Operators

```python
import math  # noqa: F401

a, b = 10, 3
print(f"a = {a}, b = {b}")
print(f"a + b = {a + b}")
print(f"a - b = {a - b}")
print(f"a * b = {a * b}")
print(f"a / b = {a / b:.4f}")
print(f"a // b = {a // b}")
print(f"a % b = {a % b}")
print(f"a ** b = {a ** b}")
```

## Comparison Operators

```python
x, y = 5, 8
print(f"{x} == {y}: {x == y}")
print(f"{x} != {y}: {x != y}")
print(f"{x} < {y}: {x < y}")
print(f"{x} > {y}: {x > y}")
print(f"{x} <= {y}: {x <= y}")
print(f"{x} >= {y}: {x >= y}")
```

## Logical Operators

```python
a = True
b = False
print(f"a and b: {a and b}")
print(f"a or b: {a or b}")
print(f"not a: {not a}")
print(f"not b: {not b}")
```

## Assignment Operators

```python
count = 10
print(f"Initial: count = {count}")
count += 5
print(f"After += 5: count = {count}")
count *= 2
print(f"After *= 2: count = {count}")
count -= 7
print(f"After -= 7: count = {count}")
count //= 3
print(f"After //= 3: count = {count}")
```

## Operator Precedence

```python
# Without parentheses
result = 5 + 3 * 2 ** 2
print(f"5 + 3 * 2 ** 2 = {result}")

# With parentheses for clarity
result = 5 + (3 * (2 ** 2))
print(f"5 + (3 * (2 ** 2)) = {result}")

# Different grouping
result = (5 + 3) * 2 ** 2
print(f"(5 + 3) * 2 ** 2 = {result}")
```

## Biotechnology Example: Drug Dosage

```python
dose_mg = 50
weight_kg = 70
half_life = 6
time_hours = 24

concentration = dose_mg / weight_kg * (0.5) ** (time_hours / half_life)
therapeutic_min = 0.1
therapeutic_max = 1.0
in_range = therapeutic_min <= concentration <= therapeutic_max

print(f"Dose: {dose_mg} mg, Weight: {weight_kg} kg")
print(f"Concentration after {time_hours}h: {concentration:.3f} mg/kg")
print(f"Therapeutic range [{therapeutic_min}, {therapeutic_max}]")
print(f"In range: {in_range}")
```

## SaaS Example: Churn Risk

```python
days_since_login = 45
support_tickets = 3
tier = "basic"

low_engagement = days_since_login > 30
multiple_issues = support_tickets >= 3
is_basic = tier == "basic"
high_risk = (low_engagement or multiple_issues) and is_basic

print(f"Low engagement: {low_engagement}")
print(f"Multiple issues: {multiple_issues}")
print(f"Basic tier: {is_basic}")
print(f"High churn risk: {high_risk}")
```

## Summary
- Arithmetic: +, -, *, /, //, %, **
- Comparison: ==, !=, <, >, <=, >= (return bool)
- Logical: and, or, not
- Assignment: +=, -=, *=, /=, //=, %=, **=
- Use parentheses to make precedence clear
