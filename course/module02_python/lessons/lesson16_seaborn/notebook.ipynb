# Lesson 16: Seaborn

Learn statistical data visualization with Seaborn.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

sns.set_theme()
tips = sns.load_dataset("tips")
print(tips.head())
```

## Scatter Plot

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="time", size="size")
plt.title("Tips by Total Bill")
plt.show()
```

## Box Plot

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=tips, x="day", y="total_bill", hue="sex")
plt.title("Bill Distribution by Day")
plt.show()
```

## Violin Plot

```python
plt.figure(figsize=(10, 6))
sns.violinplot(data=tips, x="day", y="total_bill", hue="sex", split=True)
plt.title("Bill Distribution (Violin)")
plt.show()
```

## Pairplot

```python
sns.pairplot(data=tips, hue="sex", diag_kind="kde")
plt.show()
```

## Heatmap

```python
numeric = tips.select_dtypes(include=[np.number])
corr = numeric.corr()

plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, cmap="coolwarm", center=0, square=True)
plt.title("Correlation Heatmap")
plt.show()
```

## Distribution Plot

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=tips, x="total_bill", hue="time", kde=True)
plt.title("Bill Distribution by Time")
plt.show()
```

## Biotechnology Example

```python
np.random.seed(42)
df = pd.DataFrame({
    "Gene": np.random.choice(["BRCA1", "TP53", "EGFR", "MYC", "KRAS"], 100),
    "Condition": np.random.choice(["Control", "Drug_A", "Drug_B"], 100),
    "Expression": np.random.randn(100) * 2 + 5
})

plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x="Gene", y="Expression", hue="Condition")
plt.title("Gene Expression by Condition")
plt.show()
```

## SaaS Example

```python
customers = pd.DataFrame({
    "Plan": np.random.choice(["Basic", "Pro", "Enterprise"], 200),
    "Tenure": np.random.randint(1, 60, 200),
    "Spend": np.random.uniform(10, 200, 200),
    "Churned": np.random.choice([0, 1], 200, p=[0.7, 0.3])
})

numeric = customers.select_dtypes(include=[np.number])
plt.figure(figsize=(8, 6))
sns.heatmap(numeric.corr(), annot=True, cmap="coolwarm", center=0)
plt.title("Customer Metrics Correlation")
plt.show()
```

## Summary
- Seaborn: high-level statistical visualization
- Key plots: scatter, box, violin, pairplot, heatmap
- Works directly with DataFrames
- Professional defaults with `set_theme()`
- Use `hue`, `size`, `style` for multi-variable plots
- Easily customizable with Matplotlib
