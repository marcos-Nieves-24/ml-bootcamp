# Lesson 15: Matplotlib

Learn to create and customize visualizations with Matplotlib.

```python
import matplotlib.pyplot as plt
import numpy as np
```

## Line Plot

```python
x = np.linspace(0, 10, 100)
plt.figure(figsize=(10, 6))
plt.plot(x, np.sin(x), label="sin(x)", linewidth=2)
plt.plot(x, np.cos(x), label="cos(x)", linestyle="--", color="red")
plt.title("Sine and Cosine")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

## Scatter Plot

```python
np.random.seed(42)
x = np.random.randn(100)
y = 2 * x + np.random.randn(100) * 0.5

plt.figure(figsize=(8, 6))
plt.scatter(x, y, alpha=0.6, c="steelblue", edgecolors="white", linewidth=0.5)
plt.title("Scatter Plot")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(True, alpha=0.3)
plt.show()
```

## Bar Chart

```python
genes = ["BRCA1", "TP53", "EGFR", "MYC", "KRAS"]
values = [2.5, -1.2, 3.8, 0.9, -2.1]
colors = ["green" if v > 0 else "red" for v in values]

plt.figure(figsize=(8, 6))
plt.bar(genes, values, color=colors, alpha=0.7)
plt.title("Gene Expression")
plt.xlabel("Gene")
plt.ylabel("Fold Change")
plt.axhline(y=0, color="black", linewidth=0.5)
plt.show()
```

## Histogram

```python
data = np.random.randn(1000)

plt.figure(figsize=(10, 6))
plt.hist(data, bins=30, edgecolor="white", alpha=0.7, density=True)
plt.title("Distribution")
plt.xlabel("Value")
plt.ylabel("Density")
plt.grid(True, alpha=0.3)
plt.show()
```

## Subplots

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

x = np.linspace(0, 10, 100)
data = np.random.randn(500)

axes[0, 0].plot(x, np.sin(x))
axes[0, 0].set_title("Sine")

axes[0, 1].plot(x, np.cos(x), color="red")
axes[0, 1].set_title("Cosine")

axes[1, 0].scatter(np.random.randn(50), np.random.randn(50), alpha=0.6)
axes[1, 0].set_title("Scatter")

axes[1, 1].hist(data, bins=20, edgecolor="white")
axes[1, 1].set_title("Histogram")

plt.tight_layout()
plt.show()
```

## Biotechnology Example

```python
conditions = ["Control", "Treatment_A", "Treatment_B"]
genes = ["BRCA1", "TP53", "EGFR", "MYC"]
expression = np.random.randn(4, 3) * 1.5 + 2

fig, ax = plt.subplots(figsize=(10, 6))
x_pos = np.arange(len(genes))
width = 0.25

for i, (cond, color) in enumerate(zip(conditions, ["steelblue", "coral", "seagreen"])):
    offset = (i - 1) * width
    ax.bar(x_pos + offset, expression[:, i], width, label=cond, color=color, alpha=0.8)

ax.set_xticks(x_pos)
ax.set_xticklabels(genes)
ax.legend()
ax.set_title("Gene Expression by Condition")
plt.show()
```

## Summary
- Figure: the window, Axes: the plot area
- Line, scatter, bar, histogram plots
- Customize with labels, legends, colors
- Subplots for multi-panel figures
- Save as PNG, PDF, SVG
- Always label axes and include legends
