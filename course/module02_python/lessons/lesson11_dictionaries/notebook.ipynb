# Lesson 11: Dictionaries

Learn about Python's key-value data structure — dictionaries.

## Creating and Accessing Dictionaries

```python
gene_expression = {
    "BRCA1": 2.5,
    "TP53": 1.8,
    "EGFR": 3.2,
    "MYC": 0.9,
}

print(f"Full dict: {gene_expression}")
print(f"BRCA1: {gene_expression['BRCA1']}")
print(f"TP53: {gene_expression.get('TP53')}")
print(f"Unknown: {gene_expression.get('ALK', 'N/A')}")
```

## Dictionary Methods

```python
d = {"a": 1, "b": 2, "c": 3}
print(f"Keys: {list(d.keys())}")
print(f"Values: {list(d.values())}")
print(f"Items: {list(d.items())}")

# Iterating
for key, value in d.items():
    print(f"  {key} → {value}")
```

## Dictionary Comprehension

```python
squares = {x: x ** 2 for x in range(5)}
print(f"Squares: {squares}")

names = ["Alice", "Bob", "Charlie"]
scores = [95, 87, 92]
gradebook = {n: s for n, s in zip(names, scores)}
print(f"Gradebook: {gradebook}")
```

## Counting with Dictionaries

```python
dna = "ATCGATCGATCG"
counts = {}
for base in dna:
    counts[base] = counts.get(base, 0) + 1
print(f"Base counts: {counts}")

# Using Counter
from collections import Counter
counter = Counter(dna)
print(f"Counter: {dict(counter)}")
```

## DefaultDict

```python
from collections import defaultdict

data = [("up", "BRCA1"), ("down", "TP53"), ("up", "EGFR"),
        ("down", "MYC"), ("down", "KRAS"), ("up", "ALK")]

groups = defaultdict(list)
for status, gene in data:
    groups[status].append(gene)

print(f"Groups: {dict(groups)}")
```

## Biotechnology Example

```python
codon_table = {
    "ATG": "M", "TAA": "*", "TAG": "*", "TGA": "*",
    "TTT": "F", "TTC": "F", "TTA": "L", "TTG": "L",
}

def translate(dna):
    protein = []
    for i in range(0, len(dna) - 2, 3):
        codon = dna[i:i+3].upper()
        aa = codon_table.get(codon, "?")
        protein.append(aa)
        if aa == "*":
            break
    return "".join(protein)

print(translate("ATGGCCCTTAAGTAATGA"))
```

## SaaS Example

```python
log = [
    ("user_001", "login"), ("user_002", "purchase"),
    ("user_001", "view"), ("user_003", "login"),
]

activity = {}
for user, event in log:
    if user not in activity:
        activity[user] = {}
    activity[user][event] = activity[user].get(event, 0) + 1

for user, events in sorted(activity.items()):
    print(f"{user}: {events}")
```

## Summary
- Dictionaries: key → value mappings
- Keys must be hashable (immutable)
- `get()` for safe access, `[]` for guaranteed keys
- Comprehensions: `{k: v for ...}`
- defaultdict and Counter for specialized needs
- Fast O(1) lookup
