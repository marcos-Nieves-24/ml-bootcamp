# Lesson 12: Sets

Learn about unordered collections of unique elements.

## Creating Sets

```python
empty = set()
fruits = {"apple", "banana", "cherry", "apple"}
from_list = set([1, 2, 2, 3, 3, 4])
from_string = set("hello")

print(f"Empty: {empty}")
print(f"Fruits: {fruits}")
print(f"From list: {from_list}")
print(f"From string: {from_string}")
```

## Set Operations

```python
a = {1, 2, 3, 4, 5}
b = {4, 5, 6, 7, 8}

print(f"A: {a}")
print(f"B: {b}")
print(f"Union: {a | b}")
print(f"Intersection: {a & b}")
print(f"Difference (A-B): {a - b}")
print(f"Symmetric diff: {a ^ b}")
```

## Set Methods

```python
s = {1, 2, 3}
s.add(4)
print(f"After add: {s}")
s.discard(10)  # No error
s.remove(1)
print(f"After remove: {s}")
popped = s.pop()
print(f"Popped: {popped}, Remaining: {s}")
```

## Duplicate Removal

```python
user_ids = [101, 102, 103, 101, 104, 105, 102, 106]
unique = list(set(user_ids))
print(f"Original: {user_ids}")
print(f"Unique: {sorted(unique)}")
```

## Membership Testing

```python
valid_genes = {"BRCA1", "TP53", "EGFR", "MYC", "KRAS"}
for gene in ["BRCA1", "INVALID"]:
    if gene in valid_genes:
        print(f"{gene}: VALID")
    else:
        print(f"{gene}: INVALID")
```

## Biotechnology Example

```python
exp1 = {"BRCA1", "TP53", "EGFR", "MYC", "KRAS"}
exp2 = {"TP53", "EGFR", "ALK", "ROS1", "KRAS"}
exp3 = {"BRCA1", "MYC", "ALK", "GATA2"}

common = exp1 & exp2 & exp3
print(f"Common in all: {common if common else 'None'}")

in_any_two = (exp1 & exp2) | (exp1 & exp3) | (exp2 & exp3)
print(f"In at least 2: {in_any_two}")

unique_1 = exp1 - exp2 - exp3
print(f"Unique to exp1: {unique_1}")
```

## SaaS Example

```python
basic = {"login", "view", "search", "logout"}
premium = {"login", "view", "search", "export", "api", "analytics", "logout"}

common = basic & premium
print(f"Common features: {common}")

additions = premium - basic
print(f"Premium additions: {additions}")

print(f"Basic ⊆ Premium? {basic.issubset(premium)}")
```

## Summary
- Sets: unordered, unique, mutable
- Operations: union, intersection, difference, symmetric diff
- `in` is O(1) — much faster than lists
- Frozenset: immutable, hashable version
- Great for deduplication and comparison
