# Lesson 7: Loops

Learn to repeat operations using for loops, while loops, and loop control statements.

## The for Loop

```python
genes = ["BRCA1", "TP53", "EGFR", "MYC"]
for gene in genes:
    print(f"Analyzing {gene}...")
```

## The range() Function

```python
print("range(5):", list(range(5)))
print("range(2, 8):", list(range(2, 8)))
print("range(0, 10, 2):", list(range(0, 10, 2)))
print("range(10, 0, -2):", list(range(10, 0, -2)))
```

```python
# Sum of numbers 1 to 10
total = 0
for i in range(1, 11):
    total += i
print(f"Sum 1 to 10: {total}")
```

## The while Loop

```python
count = 0
while count < 5:
    print(f"Count: {count}")
    count += 1
print("Done!")
```

## Break and Continue

```python
print("Continue demo (skip 3):")
for i in range(10):
    if i == 3:
        continue
    if i == 7:
        break
    print(i, end=" ")
print("\nBreak at 7")
```

## Nested Loops

```python
print("Multiplication table (3x3):")
for i in range(1, 4):
    for j in range(1, 4):
        print(f"{i} × {j} = {i * j}", end="\t")
    print()
```

## Enumerate and Zip

```python
# enumerate() for index and value
genes = ["BRCA1", "TP53", "EGFR"]
for i, gene in enumerate(genes):
    print(f"{i}: {gene}")

print()
# zip() for parallel iteration
expressions = [2.5, 1.8, 3.2]
for gene, expr in zip(genes, expressions):
    print(f"{gene}: {expr}")
```

## Biotechnology Example

```python
sequences = [
    "ATCGATCGATCG",
    "GGGGCCCCCAAAA",
    "ATATATATATAT",
    "CGCGCGCGCGCG",
]

for seq in sequences:
    gc = (seq.count("G") + seq.count("C")) / len(seq) * 100
    print(f"{seq}: GC = {gc:.1f}%")
```

## SaaS Example

```python
activity = [
    {"month": "Jan", "users": 1200, "churned": 45},
    {"month": "Feb", "users": 1250, "churned": 38},
    {"month": "Mar", "users": 1300, "churned": 52},
]

total_churned = 0
for m in activity:
    rate = (m["churned"] / m["users"]) * 100
    print(f"{m['month']}: {rate:.1f}% churn")
    total_churned += m["churned"]
print(f"Total churned: {total_churned}")
```

## Summary
- `for` loops iterate over sequences
- `while` loops repeat on conditions
- `range()` generates numeric sequences
- `break` exits, `continue` skips
- `enumerate()` gives index+value
- `zip()` iterates multiple sequences in parallel
