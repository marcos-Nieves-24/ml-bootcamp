## Objetivos de Aprendizaje

Al finalizar este módulo, podrás:
1. Instalar Python y configurar un entorno de desarrollo
2. Escribir código Python usando variables, tipos de datos, operadores y control de flujo
3. Usar estructuras de datos: listas, tuplas, diccionarios y conjuntos
4. Escribir código reutilizable con funciones
5. Realizar cómputo numérico con NumPy
6. Analizar datos con Pandas
7. Crear visualizaciones con Matplotlib y Seaborn
