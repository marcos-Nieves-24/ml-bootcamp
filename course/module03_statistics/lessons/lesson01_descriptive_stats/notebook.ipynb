# Lesson 1: Descriptive Statistics

## Measures of Central Tendency and Dispersion

In this notebook we will:
- Compute mean, median, mode, variance, std, range, IQR
- Identify outliers using the IQR method
- Visualize data with boxplots
- Apply these concepts to real-world datasets

## Setup

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

plt.style.use('ggplot')
%matplotlib inline
```

## 1. Central Tendency

```python
scores = np.array([78, 85, 92, 67, 88, 95, 73, 81, 90, 76,
                   84, 91, 69, 87, 93, 75, 82, 89, 96, 71,
                   79, 86, 94, 68, 83, 97, 77, 80, 99, 100])

mean = np.mean(scores)
median = np.median(scores)
mode = pd.Series(scores).mode().values

print(f"Mean: {mean:.2f}")
print(f"Median: {median:.2f}")
print(f"Mode: {mode}")
```

```python
plt.figure(figsize=(10, 3))
plt.plot(scores, np.zeros_like(scores), 'o', markersize=8, alpha=0.6)
plt.axvline(mean, color='red', linestyle='--', linewidth=2, label=f'Mean={mean:.1f}')
plt.axvline(median, color='blue', linestyle='-', linewidth=2, label=f'Median={median:.1f}')
plt.legend()
plt.title('Exam Scores with Mean and Median')
plt.xlabel('Score')
plt.tight_layout()
plt.show()
```

## 2. Dispersion

```python
variance = np.var(scores, ddof=0)
std_dev = np.std(scores, ddof=0)
data_range = np.ptp(scores)
q1 = np.percentile(scores, 25)
q3 = np.percentile(scores, 75)
iqr = q3 - q1

print(f"Variance: {variance:.2f}")
print(f"Standard Deviation: {std_dev:.2f}")
print(f"Range: {data_range}")
print(f"Q1: {q1:.2f}")
print(f"Q3: {q3:.2f}")
print(f"IQR: {iqr:.2f}")
```

## 3. Outlier Detection

```python
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr
outliers = scores[(scores < lower_bound) | (scores > upper_bound)]

print(f"Outlier bounds: [{lower_bound:.2f}, {upper_bound:.2f}]")
print(f"Number of outliers: {len(outliers)}")
print(f"Outliers: {outliers}")
```

```python
plt.figure(figsize=(8, 4))
plt.boxplot(scores, vert=False, patch_artist=True)
plt.title('Boxplot of Exam Scores')
plt.xlabel('Score')
plt.tight_layout()
plt.show()
```

## 4. Biotechnology Example: Gene Expression

```python
np.random.seed(42)
tp53_expression = np.random.normal(loc=8.5, scale=2.0, size=50)

df = pd.DataFrame({'TP53_expression': tp53_expression})
summary = df.describe()
print(summary)
```

```python
q1 = df['TP53_expression'].quantile(0.25)
q3 = df['TP53_expression'].quantile(0.75)
iqr = q3 - q1
lower = q1 - 1.5 * iqr
print(f"Samples with unusually low TP53 expression (below {lower:.2f}):")
print(df[df['TP53_expression'] < lower])
```

## 5. SaaS Example: Daily Active Users

```python
np.random.seed(42)
dau = np.random.poisson(lam=1500, size=90) + np.random.randint(-50, 50, 90)

print("Daily Active Users - Descriptive Statistics")
print(f"Mean: {np.mean(dau):.0f}")
print(f"Median: {np.median(dau):.0f}")
print(f"Std Dev: {np.std(dau, ddof=0):.0f}")
print(f"Min: {np.min(dau)}, Max: {np.max(dau)}")
```

```python
pd.Series(dau).plot(kind='hist', bins=15, edgecolor='black')
plt.title('Distribution of Daily Active Users')
plt.xlabel('Users')
plt.tight_layout()
plt.show()
```
