## Axioms, Bayes Theorem, and Random Variables

In this notebook we will:
- Simulate probability experiments
- Apply Bayes' theorem to diagnostic testing
- Work with random variables
- Visualize probability concepts

## Setup

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

plt.style.use('ggplot')
%matplotlib inline
```

## 1. Simulating Probability: Coin Flips

```python
np.random.seed(42)
n_flips = 10000
flips = np.random.choice(['H', 'T'], size=n_flips)
p_heads = np.mean(flips == 'H')
print(f"P(Heads) after {n_flips} flips: {p_heads:.4f} (theoretical: 0.5)")
```

## 2. Bayes Theorem: Medical Diagnostic Test

```python
prevalence = 0.01
sensitivity = 0.95
specificity = 0.90

p_pos_given_dis = sensitivity
p_pos_given_health = 1 - specificity
p_dis_given_pos = (p_pos_given_dis * prevalence) / (
    p_pos_given_dis * prevalence + p_pos_given_health * (1 - prevalence)
)

print(f"P(disease | positive) = {p_dis_given_pos:.3f}")
print(f"Interpretation: Only {p_dis_given_pos*100:.1f}% of positive tests"
      f" are true positives due to low prevalence.")
```

```python
# Simulation verification
n_patients = 100000
has_disease = np.random.random(n_patients) < prevalence
test_positive = np.where(
    has_disease,
    np.random.random(n_patients) < sensitivity,
    np.random.random(n_patients) < (1 - specificity)
)
empirical = has_disease[test_positive].mean()
print(f"Empirical P(disease | positive) = {empirical:.3f}")
```

## 3. Prevalence vs Posterior Probability

```python
prevalences = np.linspace(0.001, 0.5, 50)
posteriors = []

for prev in prevalences:
    posterior = (sensitivity * prev) / (
        sensitivity * prev + (1 - specificity) * (1 - prev)
    )
    posteriors.append(posterior)

plt.figure(figsize=(8, 5))
plt.plot(prevalences, posteriors, linewidth=2)
plt.axhline(0.5, color='red', linestyle='--', label='P = 0.5 (informative threshold)')
plt.xlabel('Prevalence')
plt.ylabel('P(disease | positive)')
plt.title('Posterior Probability vs Disease Prevalence')
plt.legend()
plt.tight_layout()
plt.show()
```

## 4. Random Variables: Expected Value and Variance

```python
# Discrete random variable: number of heads in 10 coin flips
n, p = 10, 0.5
binom = stats.binom(n, p)
k = np.arange(0, n+1)
probs = binom.pmf(k)

plt.figure(figsize=(8, 4))
plt.bar(k, probs, width=0.8, color='steelblue', edgecolor='black')
plt.axvline(binom.mean(), color='red', linestyle='--', label=f'Mean = {binom.mean():.1f}')
plt.xlabel('Number of Heads')
plt.ylabel('Probability')
plt.title('PMF: Number of Heads in 10 Coin Flips')
plt.legend()
plt.tight_layout()
plt.show()

print(f"Expected value (E[X]): {binom.mean():.2f}")
print(f"Variance: {binom.var():.2f}")
```
