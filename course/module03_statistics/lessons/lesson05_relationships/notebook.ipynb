# Lesson 5: Relationships Between Variables

## Covariance, Pearson and Spearman Correlation

In this notebook we will:
- Compute covariance and correlation
- Compare Pearson vs Spearman on different relationships
- Create correlation matrix heatmaps
- Analyze real datasets

## Setup

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import pearsonr, spearmanr

plt.style.use('ggplot')
%matplotlib inline
```

## 1. Generate Data with Different Relationships

```python
np.random.seed(42)
n = 200
x = np.random.normal(0, 1, n)

y_linear = 2 * x + np.random.normal(0, 0.5, n)
y_nonlinear = x**3 + np.random.normal(0, 1, n)
y_quadratic = x**2 + np.random.normal(0, 0.5, n)
y_none = np.random.normal(0, 1, n)

fig, axes = plt.subplots(2, 2, figsize=(10, 8))
axes[0, 0].scatter(x, y_linear, alpha=0.6)
axes[0, 0].set_title(f'Linear (P={pearsonr(x,y_linear)[0]:.2f}, S={spearmanr(x,y_linear)[0]:.2f})')
axes[0, 1].scatter(x, y_nonlinear, alpha=0.6)
axes[0, 1].set_title(f'Cubic (P={pearsonr(x,y_nonlinear)[0]:.2f}, S={spearmanr(x,y_nonlinear)[0]:.2f})')
axes[1, 0].scatter(x, y_quadratic, alpha=0.6)
axes[1, 0].set_title(f'Quadratic (P={pearsonr(x,y_quadratic)[0]:.2f}, S={spearmanr(x,y_quadratic)[0]:.2f})')
axes[1, 1].scatter(x, y_none, alpha=0.6)
axes[1, 1].set_title(f'No Relationship (P={pearsonr(x,y_none)[0]:.2f}, S={spearmanr(x,y_none)[0]:.2f})')
plt.tight_layout()
plt.show()
```

## 2. Correlation Matrix: Iris Dataset

```python
iris = sns.load_dataset('iris')
numeric_cols = iris.select_dtypes(include=[np.number])
corr_matrix = numeric_cols.corr()

plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='RdBu_r', center=0,
            square=True, linewidths=1, fmt='.2f')
plt.title('Iris Dataset - Correlation Matrix')
plt.tight_layout()
plt.show()
```

## 3. Penguins: Pairplot and Correlations

```python
penguins = sns.load_dataset('penguins').dropna()

plt.figure(figsize=(8, 6))
sns.heatmap(penguins.corr(numeric_only=True), annot=True, cmap='coolwarm',
            center=0, square=True, fmt='.2f')
plt.title('Penguin Measurements - Correlation Matrix')
plt.tight_layout()
plt.show()
```

```python
sns.pairplot(penguins, hue='species')
plt.show()
```

## 4. SaaS Metrics Correlation

```python
np.random.seed(42)
n_users = 500
saas_data = pd.DataFrame({
    'session_duration_min': np.random.exponential(20, n_users),
    'pages_per_session': np.random.poisson(5, n_users),
    'days_since_last_login': np.random.exponential(7, n_users),
    'support_tickets': np.random.poisson(0.5, n_users),
    'subscription_months': np.random.poisson(12, n_users)
})

plt.figure(figsize=(8, 6))
sns.heatmap(saas_data.corr(), annot=True, cmap='RdBu_r', center=0,
            square=True, fmt='.2f')
plt.title('SaaS User Metrics - Correlation Matrix')
plt.tight_layout()
plt.show()
```
