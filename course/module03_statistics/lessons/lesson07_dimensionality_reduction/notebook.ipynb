# Lesson 7: Dimensionality Reduction (PCA)

## Principal Component Analysis

In this notebook we will:
- Apply PCA to reduce dimensionality
- Interpret explained variance ratios
- Visualize high-dimensional data in 2D
- Understand feature contributions (loadings)

## Setup

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

plt.style.use('ggplot')
%matplotlib inline
```

## 1. PCA on Iris Dataset

```python
iris = sns.load_dataset('iris')
X = iris.drop('species', axis=1)
y = iris['species']

# Standardize
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# PCA
pca = PCA()
X_pca = pca.fit_transform(X_scaled)

print("Explained variance ratio:", pca.explained_variance_ratio_)
print("Cumulative:", np.cumsum(pca.explained_variance_ratio_))
```

```python
# Scree plot + 2D projection
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

axes[0].bar(range(1, 5), pca.explained_variance_ratio_, color='steelblue')
axes[0].plot(range(1, 5), np.cumsum(pca.explained_variance_ratio_), 'ro-')
axes[0].set_xlabel('PC')
axes[0].set_ylabel('Explained Variance')
axes[0].set_title('Scree Plot')

species_colors = {'setosa': 'red', 'versicolor': 'blue', 'virginica': 'green'}
for species, color in species_colors.items():
    mask = y == species
    axes[1].scatter(X_pca[mask, 0], X_pca[mask, 1], c=color, label=species, alpha=0.7)
axes[1].set_xlabel(f'PC1 ({pca.explained_variance_ratio_[0]:.1%})')
axes[1].set_ylabel(f'PC2 ({pca.explained_variance_ratio_[1]:.1%})')
axes[1].set_title('PCA 2D Projection')
axes[1].legend()
plt.tight_layout()
plt.show()
```

## 2. PCA Loadings

```python
loadings = pd.DataFrame(
    pca.components_.T,
    columns=['PC1', 'PC2', 'PC3', 'PC4'],
    index=iris.columns[:4]
)
print("PCA Loadings:")
print(loadings)

# Heatmap of loadings
plt.figure(figsize=(6, 4))
sns.heatmap(loadings, annot=True, cmap='RdBu_r', center=0)
plt.title('Feature Contributions to Principal Components')
plt.tight_layout()
plt.show()
```

## 3. PCA on Wine Dataset

```python
from sklearn.datasets import load_wine
wine = load_wine()
X_wine = wine.data
y_wine = wine.target

X_wine_scaled = StandardScaler().fit_transform(X_wine)
pca_wine = PCA(n_components=2)
X_wine_pca = pca_wine.fit_transform(X_wine_scaled)

plt.figure(figsize=(8, 6))
plt.scatter(X_wine_pca[:, 0], X_wine_pca[:, 1], c=y_wine, cmap='viridis', alpha=0.7)
plt.xlabel(f'PC1 ({pca_wine.explained_variance_ratio_[0]:.1%})')
plt.ylabel(f'PC2 ({pca_wine.explained_variance_ratio_[1]:.1%})')
plt.title('Wine Dataset PCA')
plt.colorbar(label='Cultivar')
plt.show()

print(f"Total explained variance: {pca_wine.explained_variance_ratio_.sum():.3f}")
```
