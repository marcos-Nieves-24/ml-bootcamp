# Lesson 2: Data Distribution

## Histograms, Density Plots, Skewness, and Kurtosis

In this notebook we will:
- Create histograms with different bin widths
- Plot kernel density estimates (KDE)
- Compute and interpret skewness and kurtosis
- Compare symmetric, skewed, and heavy-tailed distributions

## Setup

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import skew, kurtosis

plt.style.use('ggplot')
%matplotlib inline
```

## 1. Generate Data with Different Distributions

```python
np.random.seed(42)
symmetric = np.random.normal(loc=50, scale=10, size=1000)
right_skewed = np.random.exponential(scale=10, size=1000)
left_skewed = -np.random.exponential(scale=10, size=1000) + 100
heavy_tail = np.random.standard_t(df=3, size=1000)

fig, axes = plt.subplots(2, 2, figsize=(12, 8))
axes[0, 0].hist(symmetric, bins=25, edgecolor='black', color='steelblue')
axes[0, 0].set_title('Symmetric (Normal)')
axes[0, 1].hist(right_skewed, bins=25, edgecolor='black', color='coral')
axes[0, 1].set_title('Right-Skewed (Exponential)')
axes[1, 0].hist(left_skewed, bins=25, edgecolor='black', color='seagreen')
axes[1, 0].set_title('Left-Skewed')
axes[1, 1].hist(heavy_tail, bins=25, edgecolor='black', color='purple')
axes[1, 1].set_title('Heavy-Tailed (t-distribution)')
plt.tight_layout()
plt.show()
```

## 2. Density Plots

```python
plt.figure(figsize=(10, 5))
sns.kdeplot(symmetric, label='Symmetric', fill=True)
sns.kdeplot(right_skewed, label='Right-Skewed', fill=True)
sns.kdeplot(left_skewed, label='Left-Skewed', fill=True)
plt.title('Density Plots Comparison')
plt.legend()
plt.tight_layout()
plt.show()
```

## 3. Skewness and Kurtosis

```python
datasets = {
    'Symmetric': symmetric,
    'Right-Skewed': right_skewed,
    'Left-Skewed': left_skewed,
    'Heavy-Tailed': heavy_tail
}

print("Distribution Shape Statistics")
print("-" * 50)
for name, data in datasets.items():
    sk = skew(data)
    kt = kurtosis(data)  # excess kurtosis
    print(f"{name:20s} | Skewness: {sk:6.3f} | Kurtosis: {kt:6.3f}")
```

## 4. Real Data: Penguins Body Mass

```python
penguins = sns.load_dataset('penguins').dropna()

fig, axes = plt.subplots(1, 2, figsize=(12, 4))
axes[0].hist(penguins['body_mass_g'], bins=25, edgecolor='black', color='steelblue')
axes[0].set_title('Penguin Body Mass (Histogram)')
axes[0].set_xlabel('Body Mass (g)')

sns.histplot(penguins['body_mass_g'], kde=True, bins=25, ax=axes[1])
axes[1].set_title('Penguin Body Mass (Histogram + KDE)')
axes[1].set_xlabel('Body Mass (g)')
plt.tight_layout()
plt.show()

print(f"Skewness: {skew(penguins['body_mass_g']):.3f}")
print(f"Kurtosis: {kurtosis(penguins['body_mass_g']):.3f}")
print(f"Mean: {penguins['body_mass_g'].mean():.0f} g")
print(f"Median: {penguins['body_mass_g'].median():.0f} g")
```

## 5. Biotechnology Example: Log-Transformation

```python
np.random.seed(42)
gene_expression = np.random.lognormal(mean=2.0, sigma=0.8, size=1000)

fig, axes = plt.subplots(1, 2, figsize=(12, 4))
axes[0].hist(gene_expression, bins=30, edgecolor='black', color='steelblue')
axes[0].set_title('Raw Gene Expression')

log_transformed = np.log1p(gene_expression)
axes[1].hist(log_transformed, bins=30, edgecolor='black', color='coral')
axes[1].set_title('Log-Transformed')
plt.tight_layout()
plt.show()

print(f"Raw skewness: {skew(gene_expression):.3f}")
print(f"Log skewness: {skew(log_transformed):.3f}")
```

## 6. SaaS Example: Session Durations

```python
np.random.seed(42)
session_duration = np.random.exponential(scale=300, size=2000)

plt.figure(figsize=(10, 4))
sns.histplot(session_duration, bins=40, kde=True, edgecolor='black')
plt.axvline(np.mean(session_duration), color='red', linestyle='--',
            label=f'Mean: {np.mean(session_duration):.0f}s')
plt.axvline(np.median(session_duration), color='blue', linestyle='-',
            label=f'Median: {np.median(session_duration):.0f}s')
plt.title('Session Duration Distribution')
plt.legend()
plt.tight_layout()
plt.show()
```
