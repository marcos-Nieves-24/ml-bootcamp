## Building a Classifier

Let us implement a simple rule-based system that classifies AI systems based on keywords.

```python
def classify_ai(description):
    """
    Classify an AI system by capability and functionality
    based on keywords in the description.
    """
    desc_lower = description.lower()
    
    # Capability classification
    # All current systems are Narrow AI
    capability = "Narrow AI"
    
    # Functionality classification
    memory_keywords = ["learn", "past", "history", "track", "remember", "train",
                      "experience", "previous", "memory", "data"]
    has_memory = any(kw in desc_lower for kw in memory_keywords)
    
    if has_memory:
        functionality = "Limited Memory"
    else:
        functionality = "Reactive Machine"
    
    return capability, functionality


# Test the classifier
test_descriptions = [
    "A chess engine that evaluates the current board position",
    "A self-driving car that tracks nearby vehicles and learns from past trips",
    "A thermostat that reads temperature and turns heating on or off",
    "A language model trained on billions of text documents",
    "A spam filter that learns from user feedback"
]

print("Classification Results:")
print("-" * 60)
for desc in test_descriptions:
    cap, func = classify_ai(desc)
    print(f"System: {desc[:50]}...")
    print(f"  → Capability: {cap} | Functionality: {func}\n")
```

### Limitations of This Classifier

Our keyword-based classifier is a simple demonstration. In reality:
- The line between Reactive and Limited Memory can be blurry
- Some systems use both reactive and memory-based components
- The classifier cannot detect Theory of Mind or Self-Awareness (they do not exist yet)

This is a good example of a **rule-based expert system** — the topic of Lesson 4.
