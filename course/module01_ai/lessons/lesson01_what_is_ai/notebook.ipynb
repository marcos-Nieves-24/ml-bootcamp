## Simulación de un Agente Inteligente

Vamos a simular un agente inteligente simple: un **termostato**. Esta es la forma más simple de un agente de IA — percibe la temperatura, la compara con un punto de referencia y actúa para mantener el confort.

```python
def thermostat_agent(current_temp, set_point, threshold=1.0):
    """
    A simple thermostat agent.
    
    Parameters:
    - current_temp: current room temperature (sensor input)
    - set_point: desired temperature (goal)
    - threshold: acceptable deviation
    
    Returns:
    - action: 'heat', 'cool', or 'idle'
    """
    diff = current_temp - set_point
    
    if diff < -threshold:
        return 'heat'
    elif diff > threshold:
        return 'cool'
    else:
        return 'idle'


# Test the agent
temperatures = [15.0, 18.5, 21.0, 23.5, 28.0]
set_point = 21.0

print(f"Set point: {set_point}°C")
print(f"Threshold: ±1.0°C")
print("-" * 40)
for temp in temperatures:
    action = thermostat_agent(temp, set_point)
    print(f"Temperature: {temp}°C → Action: {action}")
```

### Interpretación de los Resultados

El agente termostato utiliza un sistema simple basado en reglas:
- **Sensor**: lectura de temperatura
- **Entorno**: la habitación
- **Actuador**: sistema de calefacción/refrigeración
- **Medida de rendimiento**: mantener la temperatura cerca del punto de referencia

Este es un **agente racional** — actúa para alcanzar un objetivo.

## Implementación de una Clase de Agente Genérico

Generalicemos el concepto de agente utilizando programación orientada a objetos.

```python
class Agent:
    """A generic intelligent agent."""
    
    def __init__(self, name, sensors, actuators, goal):
        self.name = name
        self.sensors = sensors
        self.actuators = actuators
        self.goal = goal
        self.history = []
    
    def perceive(self, observation):
        """Receive a sensor observation."""
        self.history.append(observation)
    
    def act(self, action):
        """Execute an action."""
        print(f"{self.name} performs: {action}")
    
    def describe(self):
        """Return a description of the agent."""
        sensors_str = ", ".join(self.sensors)
        actuators_str = ", ".join(self.actuators)
        return (
            f"Agent '{self.name}'\n"
            f"  Sensors: {sensors_str}\n"
            f"  Actuators: {actuators_str}\n"
            f"  Goal: {self.goal}\n"
            f"  Percepts so far: {len(self.history)}"
        )


# Create a Spotify-like recommendation agent
recommender = Agent(
    name="MusicRecommender",
    sensors=["listening_history", "likes", "skips", "playlists"],
    actuators=["song_suggestions", "playlist_generation"],
    goal="maximize user engagement"
)

print(recommender.describe())
```

## Ejemplo de Agente en Biotecnología
