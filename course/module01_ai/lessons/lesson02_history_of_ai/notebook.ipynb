## Dataset de Hitos de la IA

Creemos un conjunto de datos estructurado de los hitos de la IA y visualicémoslos.

```python
# AI milestones: (year, event, category, significance)
# Categories: 'foundation', 'winter', 'breakthrough', 'commercial'
milestones = [
    (1950, "Turing Test proposed", "foundation", "Conceptual foundation for AI"),
    (1956, "Dartmouth Conference", "foundation", "AI officially founded"),
    (1958, "Lisp programming language", "foundation", "Primary AI language for decades"),
    (1966, "ELIZA chatbot", "breakthrough", "First conversational AI"),
    (1974, "First AI Winter begins", "winter", "Funding cuts after overpromising"),
    (1980, "XCON expert system", "commercial", "Saved DEC $40M/year"),
    (1987, "Second AI Winter begins", "winter", "Expert systems market collapses"),
    (1997, "Deep Blue beats Kasparov", "breakthrough", "AI beats world chess champion"),
    (2006, "Deep learning breakthrough", "breakthrough", "Hinton trains deep networks"),
    (2012, "AlexNet wins ImageNet", "breakthrough", "Deep learning revolution begins"),
    (2016, "AlphaGo beats Lee Sedol", "breakthrough", "AI masters Go"),
    (2020, "AlphaFold solves protein folding", "breakthrough", "AI solves 50-year biology problem"),
    (2022, "ChatGPT launched", "commercial", "LLMs reach mainstream"),
]

print(f"Number of milestones: {len(milestones)}")
print("\nMilestones:")
for year, event, cat, sig in milestones:
    print(f"  {year} | {event:40s} | {cat:15s}")
```

## Visualizando la Línea de Tiempo de la IA

Creemos una línea de tiempo con código de colores que muestre los inviernos de la IA (rojo) y los avances (verde).

```python
# Define colors for each category
color_map = {
    'foundation': 'blue',
    'breakthrough': 'green',
    'winter': 'red',
    'commercial': 'orange'
}

# Create the figure
fig, ax = plt.subplots(figsize=(14, 6))

# Extract data
years = [m[0] for m in milestones]
events = [m[1] for m in milestones]
categories = [m[2] for m in milestones]
colors = [color_map[c] for c in categories]

# Draw the timeline
ax.scatter(years, [1]*len(years), c=colors, s=200, zorder=5)

# Draw the horizontal line
ax.axhline(y=1, color='gray', linestyle='-', linewidth=2, zorder=1)

# Annotate events
for i, (year, event, cat, sig) in enumerate(milestones):
    offset = 0.1 if i % 2 == 0 else -0.1
    ax.annotate(
        f"{event}",
        (year, 1 + offset),
        xytext=(0, 20 if offset > 0 else -20),
        textcoords="offset points",
        ha='center',
        fontsize=8,
        rotation=45,
        arrowprops=dict(arrowstyle='->', color='gray')
    )

# Highlight winter periods
ax.axvspan(1974, 1980, alpha=0.2, color='red', label='AI Winter')
ax.axvspan(1987, 1993, alpha=0.2, color='red')

# Formatting
ax.set_xlim(1945, 2027)
ax.set_ylim(0, 2)
ax.set_yticks([])
ax.set_xlabel("Year")
ax.set_title("AI History Timeline (1950–2025)", fontsize=14, fontweight='bold')
ax.legend(loc='upper left')

plt.tight_layout()
plt.show()
```

### Interpretando la Línea de Tiempo

Las regiones sombreadas en rojo marcan los dos inviernos de la IA:
- **1974–1980**: Después del colapso del optimismo temprano
- **1987–1993**: Después de que estalló la burbuja de los sistemas expertos

Observa el ritmo acelerado de los avances desde 2012. Esta es la **Primavera de la IA** actual.

La alternancia entre períodos rojos (invierno) y verdes (avance) ilustra el patrón del **ciclo de hype**.

## Comparando IA Simbólica vs IA Conexionista

Implementemos una demostración simple de cada enfoque.

```python
# Symbolic AI: Rule-based system for medical diagnosis
def symbolic_diagnosis(symptoms):
    """
    A simple rule-based diagnostic system.
    Rules are explicitly programmed.
    """
    rules = {
        ("fever", "cough", "fatigue"): "Common cold",
        ("fever", "headache", "stiff_neck"): "Meningitis (urgent)",
        ("fever", "rash", "joint_pain"): "Dengue fever",
        ("cough", "fever", "shortness_of_breath"): "Pneumonia",
        ("fatigue", "weight_loss", "night_sweats"): "Tuberculosis",
    }
    
    symptom_set = set(symptoms)
    for rule_symptoms, diagnosis in rules.items():
        if set(rule_symptoms).issubset(symptom_set):
            return diagnosis
    return "Unknown condition"


# Test the symbolic system
patient_symptoms = ["fever", "cough", "fatigue"]
result = symbolic_diagnosis(patient_symptoms)
print(f"Symptoms: {patient_symptoms}")
print(f"Diagnosis: {result}")

print("\n--- Problem: Missing symptom 'fatigue' ---")
patient_symptoms_2 = ["fever", "cough"]
result_2 = symbolic_diagnosis(patient_symptoms_2)
print(f"Symptoms: {patient_symptoms_2}")
print(f"Diagnosis: {result_2}  # Brittle! Missing one symptom → no match")
```

### Enfoque de IA Conexionista

Las redes neuronales aprenden patrones a partir de datos en lugar de seguir reglas explícitas. Aquí hay un clasificador lineal simple (una "neurona") que aprende de ejemplos.

```python
import numpy as np

# Simple perceptron: a basic building block of neural networks
class Perceptron:
    """A simple perceptron classifier."""
    
    def __init__(self, learning_rate=0.01):
        self.lr = learning_rate
        self.weights = None
        self.bias = 0
    
    def predict(self, X):
        """Make a prediction."""
        linear = np.dot(X, self.weights) + self.bias
        return np.where(linear >= 0, 1, 0)
    
    def train(self, X, y, epochs=10):
        """Train the perceptron on labeled data."""
        n_samples, n_features = X.shape
        self.weights = np.zeros(n_features)
        
        for epoch in range(epochs):
            errors = 0
            for i in range(n_samples):
                prediction = self.predict(X[i])
                error = y[i] - prediction
                if error != 0:
                    self.weights += self.lr * error * X[i]
                    self.bias += self.lr * error
                    errors += 1
            if errors == 0:
                break
        print(f"Trained in {epoch+1} epochs with {errors} final errors")


# Generate synthetic data: classify points above/below a line
np.random.seed(42)
X = np.random.randn(100, 2)
y = (X[:, 0] + X[:, 1] > 0).astype(int)

# Train the perceptron
p = Perceptron(learning_rate=0.1)
p.train(X, y, epochs=20)

# Evaluate
predictions = p.predict(X)
accuracy = np.mean(predictions == y)
print(f"Accuracy: {accuracy:.2%}")
```

### Diferencias Clave Demostradas

| Aspecto | IA Simbólica | IA Conexionista |
|---|---|---|
| Reglas | Programadas explícitamente | Aprendidas de datos |
| Fragilidad | Alta (falta una regla → fallo) | Baja (degradación gradual) |
| Interpretabilidad | Alta (las reglas son legibles) | Baja (los pesos son opacos) |
| Requisito de datos | Bajo (conocimiento experto) | Alto (muchos ejemplos) |
| Mantenimiento | Difícil (conflictos de reglas) | Más fácil (re-entrenar) |

```python
# Exercise 2: Improved diagnosis with partial matching
def improved_diagnosis(symptoms):
    """Rule-based diagnosis with partial matching."""
    conditions = {
        "Common cold": ["fever", "cough", "fatigue", "sneezing"],
        "Influenza": ["fever", "body_aches", "fatigue", "headache"],
        "COVID-19": ["fever", "cough", "loss_of_taste", "fatigue"],
    }
    
    symptom_set = set(symptoms)
    best_match = None
    best_score = 0
    
    for condition, cond_symptoms in conditions.items():
        matches = len(symptom_set.intersection(cond_symptoms))
        score = matches / len(cond_symptoms)
        if score > best_score:
            best_score = score
            best_match = condition
    
    return f"{best_match} (confidence: {best_score:.0%})"


# Test
print(improved_diagnosis(["fever", "cough"]))
print(improved_diagnosis(["fever", "fatigue", "headache"]))
```
